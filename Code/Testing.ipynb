```python
import cv2

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading the image
image_path = r"C:\Users\ROSHAN\Lane Detection\Images\solidWhiteCurve.jpg"

image1 = cv2.imread(image_path)
plt.imshow(image1)
```

```python
array = np.asarray(image1)
array.shape
```

```python
#Greyscaling

def grey(image):
    
    image = np.asarray(image)
    
    greyscaled_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    return plt.imshow(greyscaled_image) 

greyed = grey(image1)
greyed
```

```python
#Gauss
def gauss(greyed):
    
    #greyed = np.asarray(greyed)
    gaussed_image = cv2.GaussianBlur(greyed, (15, 15), 0)
    
    return plt.imshow(gaussed_image)

gaussed = gauss(image1)

gaussed
```

```python
#Canny Edge Detection

def canny(image):
    
    edges = cv2.Canny(image,50,150)
    
    return plt.imshow(edges)

cannied = canny(image1)
cannied
```
